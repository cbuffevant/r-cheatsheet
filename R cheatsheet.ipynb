## Vectors

Vectors are created with function `c`, passing the elements.

```r
v <- c(7, 8, 0, -1)
print(v)
```

```r
v[1] # vectors are 1-indexed!!!
```

Ranges can be used to defined vectors

```r
e <- 1:10
print(e)
```

Also, `seq` can be used to create sequences.

```r
s <- seq(1, 10, 0.2)
print(s)
```

Vectors with repeated values can be created with function `rep`.

```r
r <- rep(2, 5)
print(r)
```

Also, sequences can be repeated to construct new vectors.

```r
cc <- rep(c(1, 2), 3)
print(cc)
```

Operations element wise are kind of the default...

```r
print(r + e) # r is broadcasted to fill the bigger vector
```

```r
print(r * e)
```

```r
print(r ^ e)
```

Some other functions for vectors

| Sum     | `sum(r)`    |
| Max     | `max(r)`    |
| Sorting | `sort(r)`   |

## Loops

`for` loops are used to iterate vectors

```r
for (i in 1:10)
{
    print(i)
}
```

## Control statements

`if / else` control statement is used to execute different path of code.

```r
for (i in 1:10)
{
    if (i %% 2)
    {
        print('even')
    } else 
    {
        print('odd')
    }
}
```

## Matrixes

A matrix is initialized with a vector

```r
M  <- matrix(c(1, 5, 2, -1, 4, 0, 2, 7, 8), nrow=3, byrow=T)
print(M)
```

```r
print(dim(M))
```

In order to operate with matrixes and vectos, vectors needs to be created as matrixes or properly casted.

```r
X <- c(8, 9, 0)
x <- as.matrix(X)
print(x)
```

```r
print(M %*% x)
```

```r
x <- matrix(c(8, 9, 0), nrow=3, byrow=T)
print(M %*% x)
```

## Solving linear systems

In order to solve linear systems `Ax + b`, `solve` method should be used.

```r
A <- matrix(c(1, 5, 2, -1, 4, 0, 2, 7, 8), nrow=3, byrow=T)
x <- matrix(c(8, 9, 0), nrow=3, byrow=T)
b = A %*% x
```

```r
print(solve(A, b))
```

## Graphs

Function `plot` is used to create scatters plots.

```r
x <- seq(-10, 10, 0.1)
y <- x^2
plot(x, y)
```

```r
x <- seq(-10, 10, 0.1)
y <- x^2
plot(x, y, type='l')
```

```r
plot(x , y, main = "y = x ^ 2", xlab = "x", ylab = "y", col="green", type = 'l')
abline(1, 10, col="red")
points(1:10)
lines(1:10)
```

More than one graph can be displayed at the same time.

```r
par(mfrow=c(2,2))
plot(x, y, type = 'l', col="darkblue")
plot(x, y, pch = 20, col="chocolate")
```
